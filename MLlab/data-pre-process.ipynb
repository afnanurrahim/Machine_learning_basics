```python
minx=min(data["Age"])
maxx=max(data["Age"])
for i in data["Age"]:
    X = (i-minx)/(maxx-minx)
    data["Age"]=data["Age"].replace(i,X)
```

```python
minx=min(data["Salary"])
maxx=max(data["Salary"])
for i in data["Salary"]:
    X = (i-minx)/(maxx-minx)
    data["Salary"] =data["Salary"].replace(i,X)
```
